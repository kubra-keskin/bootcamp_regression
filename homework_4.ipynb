Ardından, yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin. Katsayılar istatistiksel olarak anlamlı mıdır? Nem ve rüzgar hızı için tahmini katsayıların işaretleri değişti mi? Tahmin edilen katsayıları yorumlayınız.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
tem_df = pd.read_csv("weatherHistory.csv")
tem_df.head(10)
```

```python
Y = tem_df["Apparent Temperature (C)"] - tem_df["Temperature (C)"]
X = tem_df[["Humidity","Wind Speed (km/h)"]]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

hedef değişken ile özellikler arasındaki ilişki:

y = 2.43 - 3.02*humidity - 0.1193*Wind_speed

Tahmin edilen katsayılar pvalue < 0.05 olduğu için istatistiksel olarak anlamlıdır.

Hisssedilen sıcaklık ile nem arasında pozitif ilişki, rüzgar hızıyla da negatif ilişki olmasını beklerdim.Ancak nem katsayısı negatif ilişki olduğunu gösteriyor, tahminlerimle aynı doğrultuda değil.

Humidity miktarındaki bir birimlik artış sıcaklığı 3.02 birim azaltmakta, rüzgar hızındaki bir birimlik artış ise 0.1193 birim azaltmaktadır.

R2:humidity ve wind speed değişkenlerinin hedef değişkenle olan varyansını göstermektedir.Yani R2 sıcaklığın % 28.8 varyans oranını humidity ve wind speed değişken ile açıklayabilir.R2 ifadesinin her zaman daha yüksek olmasını isteriz bu değer sıcaklığı acıklamak için düşük bir değerdir.

Şimdi de modele nem ve rüzgar hızı etkileşimini ekleyelim.

```python
tem_df["hum_wind_speed"] = tem_df["Humidity"]*tem_df["Wind Speed (km/h)"]

Y = tem_df["Apparent Temperature (C)"]-tem_df["Temperature (C)"] 
X = tem_df[["Humidity","Wind Speed (km/h)","hum_wind_speed"]]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

hedef değişken ile özellikler arasındaki ilişki:

y = 0.0839 + 0.1775*humidity + 0.0905*Wind_speed - 0.2971*hum_wind_speed

Katsayılar istatistiksel olarak anlamlıdır.Bu kez humidity ve wind speed değişken ile sıcaklık arasında pozitif ilişki varken yeni olusturduğumuz humidity ve wind speed etkileşimli değişen ile negatif ilişki vardır. 

R2 değerinin ise artmış olduğunu görüyoruz.Bu kez sıcaklık hedef değişkeninin %34.1 lik varyans oranını ele aldığımız 3 değişkenle açıklayabildiğimizi ortaya koyar.

```python
pd.options.display.max_columns = None

house = pd.read_csv("train.csv")
house.head()
```

```python
KitchenQual_type = {"Po":0,"Fa":1,"TA":2,"Gd":3,"Ex":4}
house["KitchenQual_type"] = house["KitchenQual"].map(KitchenQual_type)
```

```python
house2 = pd.get_dummies(house)
```

```python
house2.columns[house2.isnull().any()]
```

```python
house2.drop(['LotFrontage', 'MasVnrArea', 'GarageYrBlt'],axis=1,inplace=True)
```

### Model-1

Tüm değişkenleri modele aldığımızda cıkan sonuç: istatiksel olarak anlamsız olan birçok değişken mevcut fakat buna ragmen tüm değişkenler modelde yeraldığı için R2 değeri oldukça yüksek.

```python
Y = house2["SalePrice"]
X = house2.drop(["SalePrice"],axis=1)

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

### Model-2

Şimdi de sadece istatiksel olarak anlamlı olan değişkenler ile model olusturalım: ilk modelde istatiksel olarak anlamlı olan bazı değişkenlerin 2.modelde anlamsız olduğu görülüyor.Bunun nedeni, bir değişkeni her eklediğimiz veya çıkarttığımız zaman, model spesifikasyonu değişir. Bu, katsayılar için farklı tahminler ile sonuçlanır. Aynısı, katsayıların istatistiksel anlamlılığı için de geçerlidir.örneğin RoofStyle_Shed değişkeni ilk model için pvalue=0.023 ile anlamlıyken 2.modelde 0.869 değeri ile anlamsız olmustur.Bunun yanında ilk modelde katsayısı 7.737e+04 iken 2.modelde -3453.7212 olmustur.

```python
gecerli_p=list(results.pvalues[results.pvalues < 0.05].index)
```

```python
Y = house2["SalePrice"]
X = house2[gecerli_p]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

### Model-3

bizim için anlamlı olduğunu düşündüğümüz modelimizi kuralım ve farklılıkları görmek için de etkisini olmadığını düşündüğümüz değişkenlerden de ekleyelim

```python
Y = house2["SalePrice"]
X = house2[['KitchenQual_type','OverallQual','GrLivArea','GarageCars','TotalBsmtSF','YearBuilt',
            "BsmtHalfBath","EnclosedPorch","MiscVal","MSZoning_RH","Street_Grvl",
            "LotShape_IR1","Utilities_NoSeWa","Neighborhood_Blmngtn","Alley_Grvl","Heating_Floor"]]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

### Model-4

Son model için değişken seçimi:

```python
y = house2['SalePrice']
x = house2.drop("SalePrice",axis=1)

from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
feature_select = SequentialFeatureSelector(LinearRegression(),
                                          k_features = 20,
                                          floating = True,
                                          forward = True,
                                          scoring = "r2",
                                          cv = 0)
feature_select.fit(x,y)
list(feature_select.k_feature_names_)
```

model için önemli olduğunu düşündüğümüz değişkenlerle modeli tekrar calıştırırsak R2 değerinin arttığını görüyoruz ve Pvalue değeri 0.05 ten büyük olan değer olmadığı için seçtiğimiz değişkenler model için anlamlıdır.Ev fiyatları üzerinde belirgin etkileri vardır.

```python
Y = house2["SalePrice"]
X = house2[list(feature_select.k_feature_names_)]

X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```
